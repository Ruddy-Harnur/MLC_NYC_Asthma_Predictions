## 4) Join the data checked

```python
dfClean = pd.DataFrame()
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Asthma Hospitalizations (Adults) by NTA')&(dfNTA['measure']=='Average Annual Rate')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Children under 5 years old in Poverty')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Children under 6 years with elevated blood lead levels by NTA')&(dfNTA['measure']=='Rate  BLL>=5 µg/dL')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Foreign-Born')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='High School Graduation')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Limited English')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Owner-Occupied Homes')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Poverty')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Pre-1950 Homes')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Rent Burdened Households')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Unemployment')&(dfNTA['measure']=='Percent')])
dfClean = dfClean.append(dfNTA[(dfNTA['name']=='Walking Distance to a Park')&(dfNTA['measure']=='Percentage')])
```

## 5) Check data to be used (measure)

```python
#dfNTA[(dfNTA['measure']=='Percent Asian Alone')]
#dfNTA[(dfNTA['measure']=='Percent Black Alone')]
#dfNTA[(dfNTA['measure']=='Percent Hispanic Alone')]
#dfNTA[(dfNTA['measure']=='Total Population')]
#fNTA['measure'].unique()
```

## 6) Join separated data

```python
dfTransition = pd.DataFrame()
dfTransition = dfTransition.append(dfNTA[(dfNTA['measure']=='Percent Asian Alone')])
dfTransition = dfTransition.append(dfNTA[(dfNTA['measure']=='Percent Black Alone')])
dfTransition = dfTransition.append(dfNTA[(dfNTA['measure']=='Percent Hispanic Alone')])
dfTransition = dfTransition.append(dfNTA[(dfNTA['measure']=='Percent White Alone')])
dfTransition = dfTransition.append(dfNTA[(dfNTA['measure']=='Total Population')])
dfTransition['name'] = dfTransition['measure']
```

## 7) Join data

```python
dfClean = dfClean.append(dfTransition)
dfClean['name'].unique()
```

## 8) Create pivot table

```python
df_analysis_pivot = pd.pivot_table(dfClean, index='geo_place_name',values='data_value',columns='name', aggfunc=lambda x: mean(x))
df_analysis_pivot['NTA'] = df_analysis_pivot.index
df_analysis_pivot.head()
```

## 9) Merge with Occupations dataset

```python
df_occupations = pd.read_csv('InfoshareData.csv')
df_occupations.columns
```

```python
df_analysis_pivot_merged = df_analysis_pivot.merge(df_occupations, left_on='geo_place_name', right_on='Area Name')
```

```python
df_analysis_pivot_merged.insert(0, 'NTA_name', df_analysis_pivot_merged.NTA)
```

```python
df_analysis_pivot_merged
```

#### Normalize occupations

```python
list_columns_to_normalize = list(df_occupations.columns)
list_columns_to_normalize.remove('MapID')
list_columns_to_normalize.remove('Area Name')
list_columns_to_normalize
```

```python
## DIVIDE EVERYTHING BY POPULATION *100 (ok) 
for i in range(len(df_analysis_pivot_merged)):
    for column in list_columns_to_normalize:
        df_analysis_pivot_merged[column].values[i] = (float(df_analysis_pivot_merged[column].values[i])/float(df_analysis_pivot_merged['Total Population'].values[i]))*100
```

```python
## CHECK RESULTS
df_analysis_pivot_merged.head()
```

```python
df_analysis_pivot_merged.dropna(inplace=True)
```

```python
df_analysis_pivot_merged.drop(columns=['NTA'], inplace=True)
df_analysis_pivot_merged.drop(columns=['MapID','Area Name'], inplace=True)
df_analysis_pivot_merged.columns
```
